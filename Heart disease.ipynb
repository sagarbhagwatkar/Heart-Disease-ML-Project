### Let's see explanation of each of the columns:
* age: age in years

* sex: sex

1 = male

0 = female

* cp: chest pain type
Value 1: typical angina

Value 2: atypical angina

Value 3: non-anginal pain

Value 4: asymptomatic

* trestbps: resting blood pressure (in mm Hg on admission to the hospital)

* chol: serum cholestoral in mg/dl

* fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

* restecg: resting electrocardiographic results

Value 0: normal

Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)

Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

*thalach: maximum heart rate achieved

* exang: exercise induced angina (1 = yes; 0 = no)

* oldpeak: ST depression induced by exercise relative to rest

* slope: the slope of the peak exercise ST segment

Value 1: upsloping Value 2: flat Value 3: downsloping

* ca: number of major vessels (0-3) colored by flourosopy

* thal: A blood disorder called thalassemia(This factor is heredetory)

3 = normal

6 = fixed defect

7 = reversable defect
 
* target: Heart disease (0 = no, 1 = yes)

## ENSEMBLE TECHNIQUES

## Bagging classifier

```python
from sklearn.ensemble import BaggingClassifier
```

```python
DF_bagging=DataFrame(columns=['Base_Estimator','Training_Score','Testing_Score','Case_of','Difference(Training-Testing)'])
to_bag=[dt_classifier,rf_classifier,knn_classifier,svm_classifier,log_reg_classifier]
algorithm_name=['Decision_Tree','Random_Forest','KNN','Support_Vector_Classifier','Logestic_regression']
for i in range(len(to_bag)):
    case_of=''
    Difference=0
    bag_classifier = BaggingClassifier(to_bag[i],bootstrap=True,random_state=0)
    bag_classifier.fit(x_train,y_train)
    if bag_classifier.score(x_train,y_train)>bag_classifier.score(x_test,y_test):
        case_of='Overfitting'
    else:
        case_of='Underfitting'
    Difference = bag_classifier.score(x_train,y_train) - bag_classifier.score(x_test,y_test)
    DF_bagging.loc[i]=[algorithm_name[i],bag_classifier.score(x_train,y_train),bag_classifier.score(x_test,y_test),case_of,Difference]
DF_bagging
```

# Boosting classifier

## 1.Adaboost classifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
DF_adaboost=DataFrame(columns=['Base_Estimator','Training_Score','Testing_Score','Case_of','Difference(Training-Testing)'])
to_boost = [dt_classifier,rf_classifier,svm_classifier,log_reg_classifier]
algorithm_name = ['Decision_Tree','Random_Forest','Support_Vector_Classifier','Logestic_regression']
for i in range(len(to_boost)):
    case_of=''
    Difference=0
    adb_classifier = AdaBoostClassifier(to_boost[i],n_estimators=100,algorithm='SAMME',random_state=7)
    adb_classifier.fit(x_train,y_train)
    if adb_classifier.score(x_train,y_train)>adb_classifier.score(x_test,y_test):
        case_of='Overfitting'
    else:
        case_of='Underfitting'
    Difference = adb_classifier.score(x_train,y_train)-adb_classifier.score(x_test,y_test)
    DF_adaboost.loc[i]=[algorithm_name[i],adb_classifier.score(x_train,y_train),adb_classifier.score(x_test,y_test),case_of,Difference]
DF_adaboost
```

## 2.Gradiant boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
DF_gboost=DataFrame(columns=['Base_Estimator','Training_Score','Testing_Score','Case_of','Difference(Training-Testing)'])
to_boost=[dt_classifier,rf_classifier,knn_classifier,svm_classifier,log_reg_classifier]
algorithm_name=['Decision_Tree','Random_Forest','KNN','Support_Vector_Classifier','Logestic_regression']
for i in range(len(to_boost)):
    case_of=''
    Difference=0
    Gboost_classifier=GradientBoostingClassifier(n_estimators=200,max_depth=2,learning_rate=0.2,min_samples_split=8,criterion='mse')
    Gboost_classifier.fit(x_train,y_train)
    if Gboost_classifier.score(x_train,y_train)>Gboost_classifier.score(x_test,y_test):
        case_of='Overfitting'
    else:
        case_of='Underfitting'
    Difference=Gboost_classifier.score(x_train,y_train)-Gboost_classifier.score(x_test,y_test)
    DF_gboost.loc[i]=[algorithm_name[i],Gboost_classifier.score(x_train,y_train),Gboost_classifier.score(x_test,y_test),case_of,Difference]
DF_gboost
```
